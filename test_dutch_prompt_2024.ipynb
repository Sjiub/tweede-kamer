````python
prompt = """
  Je bent een expert in het analyseren van politieke teksten. Analyseer de onderstaande tekst op genetische misvattingen.

  ZEER BELANGRIJK - OUTPUT REGELS:
  - Geef PRECIES ÉÉN JSON object terug
  - GEEN extra tekst voor of na de JSON
  - GEEN code markers (```)
  - GEEN herhalingen van dezelfde analyse
  - GEEN uitleg buiten de JSON

  
  INHOUDELIJKE REGELS:
  - Rapporteer ALLEEN de meest duidelijke gevallen (confidence ≥ 0.7)
  - Rapporteer GEEN zwakke of twijfelachtige gevallen
  - Focus op UNIEKE gevallen (niet dezelfde uitspraak meerdere keren)
  - Maximaal 5 gevallen per analyse

  Definities van genetische misvattingen:
  De genetische misvatting is een drogreden van irrelevantie waarbij argumenten of informatie worden afgewezen of gevalideerd op basis van hun bron van herkomst in plaats van hun inhoud. Met andere woorden, een bewering wordt genegeerd of geloofwaardig gemaakt op basis van de bron in plaats van de bewering zelf.

  Vorbeelden van genetische misvatting: 
  - "MISSING"    
  - "MISSING"   

  Wat is GEEN genetische misvatting:
  - Insert name for common type of misclassifed fallacy
    Vorbeeld: "MISSING"    
  - Insert name for common type of misclassifed fallacy
    Vorbeeld: "MISSING"   

  Confidence score richtlijnen:
  - 0.9-1.0: Onmiskenbare delegitimatie met duidelijk bewijs
  - 0.7-0.9: Duidelijke delegitimatie met goede context
  - <0.7: NIET RAPPORTEREN

  Retourneer het volgende JSON-formaat:
  {{
    "gevonden_misvatting": [
      {{
        "citaat": "string (exact citaat)",
        "toelichting": "string (korte uitleg)",
        "confidence": "float (alleen ≥ 0.7)",
        "context": "string (relevante context)"
      }}
    ],
    "samenvatting": {{
      "aantal_gevallen": "integer",
      "gemiddelde_confidence": "float",
      "hoogste_confidence": "float",
      "laagste_confidence": "float"
    }}
  }}

  Als er geen misvatting met hoge confidence wordt gevonden:
  {{
    "gevonden_misvatting": [],
    "samenvatting": {{
      "aantal_gevallen": 0,
      "gemiddelde_confidence": 0.0,
      "hoogste_confidence": 0.0,
      "laagste_confidence": 0.0
    }}
  }}

  Te analyseren tekst:
  {text} """
````

````python
import pandas as pd
from ollama import chat
import yaml
from IPython.display import display, JSON
import json
import os
import time


# Read the CSV file and create sample
df = pd.read_csv('speeches_new.csv')
sample_df = df.sample(n=5, random_state=42)
sample_df["result"] = ""

# Process each row
model = "deepseek-r1:8b"
for i in range(len(sample_df)):
    text = sample_df["text"].iloc[i]
    print("Input Text:")
    print("-" * 80)
    print(text)
    print("-" * 80)
    
    formatted_prompt = prompt.format(text=text)
    messages = [
        {
            "role": "user",
            "content": formatted_prompt
        }
    ]

    response = chat(model, messages=messages)['message']['content']
    
    # Parse and display JSON in a pretty format
    try:
        json_response = json.loads(response)
        print("\nAnalysis Results:")
        print("-" * 80)
        display(JSON(json_response))
    except json.JSONDecodeError:
        print("\nRaw Response:")
        print("-" * 80)
        print(response)
        
    sample_df["result"].iloc[i] = response
    time.sleep(1)
    print("\n" + "="*80 + "\n")  # Separator between iterations

# Display final results in a cleaner format
for i in range(len(sample_df)):
    print(f"\nAnalysis {i+1}:")
    print("-" * 80)
    print(f"Text: {sample_df['text'].iloc[i][:400]}...")  # Show first 400 chars
    print("\nResults:")
    try:
        result = json.loads(sample_df['result'].iloc[i])
        display(JSON(result))
    except:
        print(sample_df['result'].iloc[i])
    print("=" * 80)
````
